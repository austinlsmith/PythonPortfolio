```python
Platform_group = data.groupby('Platform').Global_Sales.sum()
excel_table = pd.pivot_table(data[data.Platform.isin(Platform_group.sort_values(ascending=False)[:20].index)],values=['Global_Sales'],index=['Year'],columns=['Platform'],aggfunc='sum',margins=False)
excel_table
```

```python
plt.figure(figsize=(19,15))
sns.heatmap(excel_table['Global_Sales'],linewidths=.5,annot=True,vmin=0.01,cmap='coolwarm')
```

```python
Genre_group = data.groupby('Genre').Global_Sales.sum()
excel_table_genre = pd.pivot_table(data[data.Genre.isin(Genre_group.sort_values(ascending=False)[:20].index)],values=['Global_Sales'],index=['Year'],columns=['Platform'],aggfunc='sum',margins=False)
excel_table_genre
```

```python
plt.figure(figsize=(19,15))
sns.heatmap(excel_table_genre['Global_Sales'],linewidths=.5,annot=True,vmin=0.01,cmap='coolwarm')
```
